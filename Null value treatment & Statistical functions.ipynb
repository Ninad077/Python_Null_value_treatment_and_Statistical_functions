Null value treatment

```python
# The most important thing in Data preprocessing is to treat the null values before implementing the data on Machine 
# Learning algorithms. Because one can't apply data with missing values on Machine Learning algorithms.
# To show an example for the same we have removed some data from the "Salaries.csv" file and updated it to 
# "Salaries_new.csv". So this new file now has null entries as well.

data_new= pd.read_excel("Salaries_new.xlsx")
```

```python
data_new
```

```python
# isnull()
# isnull() function substitutes 'True' in the place where we've null values and 'False' in the place of non-null values

data_new.isnull()
```

```python
# isnull().sum()
# isnull().sum() returns the total number of null values in a particular column

data_new.isnull().sum()
```

```python
# fillna()
# fillna() is used to fill the null values with the value specified inside it's argument.

# Looking at the o/p above we have got null values for the column phd and service. Let us try to replace the null values 
# the phd column with it's mean. The query would look like

data_new["phd"].fillna(data_new["phd"].mean())
```

```python
# Now adding an extra argument inplace = True, so inplace will substitute the null values with the specified value 
# (i.e. mean in this example) not only in phd column, but also in the original dataset. 
# So if we fetch the original datset back, the eliminated null values would be reflected in the table.

data_new["phd"].fillna(data_new["phd"].mean(),inplace =True)
```

```python
data_new
```

```python
# Checking the null values for phd column now

data_new.isnull().sum()
```

```python
# As one can clearly see, the null values of phd column are eliminated (since, sum of phd null values =0)
```

```python
# Similary treating the null values of service column by it's mode

data_new["service"].fillna(data_new["service"].mode()[0],inplace=True)
```

```python
data_new.isnull().sum()
```

```python
#Task: Write a code treating null values with mode if the column is categorical (dtype: object)
# and treating null values with mean if the column is numeric (dtype: int64 ordtype: int64 ) for the source "data_new"

data_new= pd.read_excel("Salaries_new.xlsx")
```

```python
def Null_Value_Treatment(data_new):
    for i in data_new.columns:
        if data_new[i].dtypes == "int64" or data_new[i].dtypes == "float64":
            data_new[i].fillna(data_new[i].mean(),inplace=True)
        else:
            data_new[i].fillna(data_new[i].mode()[0],inplace=True)
    return data_new.isnull().sum()
        
```

```python
Null_Value_Treatment(data_new)
```
